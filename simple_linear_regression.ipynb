```python
 #import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model

#read data
data=pd.read_csv('housedata.csv')
#data.plot(kind='scatter',x='sqft_living',y='price')
#plt.title('HOUSING PRICES')
#plt.show()

#create dataframes
SQFT=pd.DataFrame(data['sqft_living'])
PRICE=pd.DataFrame(data['price'])

#build a linear model
lm=linear_model.LinearRegression()
model=lm.fit(SQFT,PRICE)

print("Coefficient: ",model.coef_)
print("Intercept: ",model.intercept_)
print("Score: ",model.score(SQFT,PRICE))

#predicted values for new inputs
SQFTnew=np.array(11111)
SQFTnew=np.atleast_2d(SQFTnew)
PRICEpredicted=model.predict(SQFTnew)
print();print()
print("New value for SQFT: ",SQFTnew)
print("Price predicted: ",PRICEpredicted)

X=pd.DataFrame([3200,10000,1020])
Y=model.predict(X)
Y=pd.DataFrame(Y)
df=pd.concat([X,Y], axis=1,keys=['New value','Predicted value'])
print();print()
print(df)

#visualization of data
data.plot(kind='scatter',x='sqft_living',y='price')
plt.xlabel('Sqft living')
plt.ylabel('Price')
plt.title('HOUSING PRICES')
plt.plot(SQFT,model.predict(SQFT),color='r')
plt.plot(SQFTnew,model.predict(SQFTnew),color='k',marker='*',markersize=8)
plt.plot(X,model.predict(X),color='y',linewidth=2)
plt.show()
```
